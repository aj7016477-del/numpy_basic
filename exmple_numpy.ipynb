```python
# Q.2 Create a 2 × 3 NumPy array and print:
#  shape , size , Number of dimensions
```

```python
a = np.arange(1 ,7).reshape(2,3)
print(a)
print("shape :" , np.shape(a))
print("size :" , np.size(a))
print("Dimensions: " ,  np.ndim(a))
```

```python
# Q.5 print 50 , first row and last column

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print(arr)
print(arr[1,1]) # print 50
print(arr[0,:]) # first row
print(arr[ : ,2]) # last column
print(arr[0:2 , 1:3])
```

```python
# Q.6 Mathematical Operations perform a+b , a-b and a*b .

a = np.array([2, 4, 6])
b = np.array([1, 3, 5])

print(a)
print(b)

print(a + b)
print(a - b)
print(a * b)
```

```python
# Q.7 using Statistical Functions print Maximum marks  , Minimum marks , 
#  Sum of all marks and Average marks
```

```python
marks = np.array([65, 70, 80, 90, 85])
print(marks)
print(np.max(marks))       # Maximum marks
print(np.min(marks))       # Minimum marks
print(np.sum(marks))       # Sum of marks
print(np.average(marks))   # Average marks
```

```python
# Q.8 using Broadcasting Multiply every element by 5 using NumPy.
```

```python
a  = np.array([[1, 2, 3],
                [4, 5, 6]])
b = np.array([5])

print(a)
print(b)
print(a * b)
```

```python
# Q.9 5×5 matrix with last row, first column, middle 3×3 ?

x = np.arange(1,26).reshape(5,5)
print(x)

# Last row
print(x[4,:])

# First column
print(x[:,0])

# Middle 3x3 matrix (top-left 3x3)
print(x[0:3, 0:3])
```
